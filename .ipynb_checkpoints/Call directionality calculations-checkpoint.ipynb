### We follow Mellow and Kärkkäinen 2005 and use the first term of eqn 92, which is the piston in an infinite baffle:

$D(\theta)\ = J_1(ka\ sin(\theta))/ ka\ sin(\theta) $

```python
%matplotlib inline
a = 15 *10.0**-3 / 2.0# range of a from Jakobsen et al. 
freqs = np.linspace(96,20,30)*10**3

ka_sweep = np.array( [ get_ka(each_freq,a) for each_freq in freqs])

def D_theta(theta,ka):
    numerator = 2*mpmath.besselj(1,ka*np.sin(theta))
    denominator = ka*np.sin(theta)
    
    dtheta = numerator/denominator
    
    return(dtheta)

angles = np.linspace(np.pi/2,-np.pi/2,500)

call_shape = np.zeros((freqs.size,angles.size))

for each_row in range(freqs.size):
    call_shape[each_row,:] = np.array([ D_theta(each_angle,ka_sweep[each_row]) for each_angle in angles    ])

dB_callshape = 20*np.log10(call_shape)

plt.figure(123,figsize=(15,12))
ax = plt.subplot(111, projection='polar')

for each_row in range(0,freqs.size,5):
    ax.plot(angles,dB_callshape[each_row,:],'*-',label='f(KHz) : '+str(freqs[each_row]/10**3));
ax.set_rticks(np.arange(0,-70,-6));
ax.set_theta_zero_location("N");plt.legend()
plt.ylabel('dB ref 0 degrees',fontsize=15);plt.title('$D(\\theta)\ of\ Piston\ in\ infinite\ baffle$',fontsize=20)
```

### Having this information, we can then go back and calculate the direction based source level of a bat  :

```python
SL = 100 # dB SPL @ 1m 

SL_theta = SL + dB_meancallshape

plt.figure(123,figsize=(15,12))
ax = plt.subplot(111, projection='polar')
ax.plot(angles,SL_theta,'*-',label='Call average source level');
ax.set_rticks(np.arange(SL,0,-12));plt.title('Direction vs Source Level, for a bat call',fontsize=20)
ax.set_theta_zero_location("N");plt.legend()
plt.text(20,190,'Piston radius $k$ = %0.02f m'%(a),fontsize = 14)
```

## Having said this, let's now move onto calculating received levels over a range of distances and angles: 



```python
def calc_RL(SL=120.0,R0=0.1,R=1.0,alpha=0):
    '''
    calculate received level based on the sonar equation
    SL : float. source level in dB SPL
    R0 : float. reference distance, at which the source level measurement was made. 
        defaults to 10cm because we're dealing with bats. 
    R : float. distance at which the receiver is from source
    alpha: float<0. medium absorption coefficient in dB/m
    '''
       
    RL = SL - 20*np.log10(R/float(R0)) + R*alpha
    
    return(RL)
```

```python
# let's now calculate the RL over a range of distances and angles : 
receiver_distance = np.linspace(0.1,3,5)
RL_rtheta = np.zeros(( receiver_distance.size,angles.size))

for i,each_distance in enumerate(receiver_distance):    
    RL_rtheta[i,:] = np.array([ calc_RL(SL_eachangle,1.0,each_distance,0) for SL_eachangle in SL_theta] )
```

```python
plt.figure(124,figsize=(15,12))
ax = plt.subplot(111, projection='polar')

for i,each_distance in enumerate(receiver_distance):
    ax.plot(angles,RL_rtheta[i,:],'*-',label='%0.02f'%each_distance);
    
ax.set_rticks(np.arange(SL,0,-12));plt.title('Direction vs Received Level, for a bat call',fontsize=20)
ax.set_theta_zero_location("N");plt.legend(title='Receiver distance (m)')
plt.text(20,190,'Piston radius $k$ = %0.02f m'%(a),fontsize = 14)
```
