```python
theta = np.arange(0,360)
hear_dirn = 2* (np.cos(np.deg2rad(theta)) -1)
call_dirn = 7 * (np.cos(np.deg2rad(theta)) -1)
```

```python
# thanks https://stackoverflow.com/questions/29394552/move-radial-tick-labels-on-a-polar-plot-in-matplotlib
```

```python
plt.figure(figsize=(8,8))
ax =plt.subplot(111, projection='polar') 
ax.set_theta_zero_location("N")
plt.plot(np.deg2rad(theta),hear_dirn, label='Hearing directionality')
plt.plot(np.deg2rad(theta),call_dirn, label='Call directionlality')
plt.xticks(np.deg2rad(np.arange(0,360,45)), ['0','45$^{\circ}$','90$^{\circ}$',
                                             '135$^{\circ}$','$\pm$180$^{\circ}$',
                                             '-135$^{\circ}$','-90$^{\circ}$',
                                             '-45$^{\circ}$'], fontsize=12)
plt.yticks(np.arange(-4,-16,-4),fontsize=12)
plt.text(np.deg2rad(5),-1,'Relative intensity, dB', fontsize=12, rotation=90)
plt.text(np.deg2rad(-13),0.5,'Off-axis angle, $^{\circ}$', fontsize=12, rotation=-24)

ax.set_rlabel_position(0);plt.legend(fontsize=12)
ax.grid(True)
```
