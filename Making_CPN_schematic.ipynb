```python
from __future__ import division
import matplotlib.pyplot  as plt
import matplotlib.image as mpimg
from matplotlib.patches import Patch
import numpy as np 
```

```python
%matplotlib notebook
```

```python
np.random.seed(1521)
```

```python
one_bat = mpimg.imread('plots/single_bat.png')
```

```python
for k in range(10)    :
    # arrows representing masker calls
    r_start, r_end  = 950, 400
    theta = np.unique( np.random.choice(np.linspace(0,2*np.pi, 30), 20) )
    x_start, x_end = r_start*np.cos(theta), r_end*np.cos(theta)
    y_start, y_end = r_start*np.sin(theta), r_end*np.sin(theta)

    # arrows representing echoes
    r_echo_s, r_echo_e = 400, 200
    theta_echo = np.linspace(np.pi/3, 2*np.pi/3, 5, endpoint=True)
    x_echo_s , y_echo_s = r_echo_s*np.cos(theta_echo), r_echo_s*np.sin(theta_echo)
    x_echo_e , y_echo_e = r_echo_e*np.cos(theta_echo), r_echo_e*np.sin(theta_echo)
    fig = plt.figure(figsize=(10,10))
    ax = plt.subplot(111)
    min_xylims, max_xylims = -1000, 1000
    plt.xlim(min_xylims,max_xylims)
    plt.ylim(min_xylims,max_xylims)
    # plot call arrows
    for x_s,y_s,x_e,y_e in zip(x_start,y_start,x_end,y_end):
        lwd = np.random.choice(np.arange(6,15,0.5),1)
        alpha_val = ((lwd-6)/9)*0.8 + 0.2# scale the alpha to the lwd
        plt.arrow(x_s, y_s, -x_e, -y_e, head_width=0.1, linewidth=lwd,
                  fc="k", ec="k",alpha=alpha_val, length_includes_head=False, overhang=-1.2)

    # plot echo arrows:
    for echo_xs, echo_ys, echo_xe, echo_ye in zip(x_echo_s, y_echo_s, x_echo_e, y_echo_e):
        plt.arrow(echo_xs, echo_ys, -echo_xe, -echo_ye, linewidth=np.random.choice(np.arange(2,8),1),
                  fc="k", ec="g",alpha=0.4, overhang=-2)

    # make custom legend here:
    legend_elements = [Patch(facecolor='grey', edgecolor='grey', label='conspecific masker call'),
                       Patch(facecolor='green', edgecolor='green', label='echoes')]

    ax.legend(handles=legend_elements, fontsize=15, loc=2, frameon=False)
    ax.axis('off')
    newax = fig.add_axes([0.5-one_bat.shape[1]/2000, 0.5-one_bat.shape[1]/2000.0,
                          one_bat.shape[0]/1000, one_bat.shape[1]/1000])
    newax.imshow(one_bat)
    newax.axis('off')

    rand_num = str(np.random.choice(np.arange(1,10000,1)))
    plt.savefig('plots/plots_round2/CPN_schematic'+'_'+rand_num+'_.png',dpi=400,pad_inches = 0, bbox_inches='tight')
```
